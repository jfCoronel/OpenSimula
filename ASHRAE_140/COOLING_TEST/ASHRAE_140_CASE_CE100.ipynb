```python
system = pro.component("system")
sys_df=system.variable_dataframe()
space_df = pro.component("space_1").variable_dataframe()
sys_df["time_frac"] = sys_df["F_load"]/sys_df["efficiency_degradation"]
sys_df["power_vi"] = sys_df["indoor_fan_power"]
sys_df["power_ve"] = sys_df["time_frac"]*108
sys_df["power_comp"] = sys_df["power"]-sys_df["power_vi"]-sys_df["power_ve"]
sys_df["Sensible_space_load"] = space_df["system_sensible_heat"]
sys_df["Latent_space_load"] = space_df["system_latent_heat"]
sys_df["Total_space_load"] = sys_df["Latent_space_load"]+sys_df["Sensible_space_load"]
sys_df["abs_humidity"] = space_df["abs_humidity"]
sys_df["COP"] = -sys_df["Total_space_load"]/sys_df["power"]

mes_df_sum = sys_df.resample("M", on='date').sum()
mes_df_mean = sys_df.resample("M", on='date').mean()
mes_df_max = sys_df.resample("M", on='date').max()
mes_df_min = sys_df.resample("M", on='date').min()

excel = pd.DataFrame()
excel["Total_consumption"] = [mes_df_sum["power"][1]/1000]
excel["Compressor_consumption"] = [mes_df_sum["power_comp"][1]/1000]
excel["Indoor_fan_consumption"] = [mes_df_sum["power_vi"][1]/1000]
excel["Outdoor_fan_consumption"] = [mes_df_sum["power_ve"][1]/1000]
excel["Total_coil_load"] = [mes_df_sum["Q_total"][1]/1000]
excel["Sensible_coil_load"] = [mes_df_sum["Q_sensible"][1]/1000]
excel["Latent_coil_load"] = [mes_df_sum["Q_latent"][1]/1000]
excel["Total_space_load"] = [-mes_df_sum["Total_space_load"][1]/1000]
excel["Sensible_space_load"] = [-mes_df_sum["Sensible_space_load"][1]/1000]
excel["Latent_space_load"] = [-mes_df_sum["Latent_space_load"][1]/1000]

excel["COP"] = [mes_df_mean["COP"][1]]
excel["T_idb"] = [mes_df_mean["T_idb"][1]]
excel["w_i"] = [mes_df_mean["abs_humidity"][1]/1000]

excel["COP_max"] = [mes_df_max["COP"][1]]
excel["T_idb_max"] = [mes_df_max["T_idb"][1]]
excel["w_i_max"] = [mes_df_max["abs_humidity"][1]/1000]

excel["COP_min"] = [mes_df_min["COP"][1]]
excel["T_idb_min"] = [mes_df_min["T_idb"][1]]
excel["w_i_min"] = [mes_df_min["abs_humidity"][1]/1000]

excel.to_clipboard()
excel
```
