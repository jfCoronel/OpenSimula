```python
system = pro.component("system")
sys_df=system.variable_dataframe()
sys_df
```

```python
space_df=pro.component("space_1").variable_dataframe()
space_df
```

```python
import pandas as pd
import psychrolib as sicro
sicro.SetUnitSystem(sicro.SI)

print("Q_heat_zone: ", space_df["system_sensible_heat"][8759]/1000)
print("Q_heat_system: ", sys_df["Q_sensible"][8759]/1000)

p = system.props["ATM_PRESSURE"]

T_odb = sys_df["T_odb"][8759]
T_owb = sys_df["T_owb"][8759]
w_o = sicro.GetHumRatioFromTWetBulb(T_odb,T_owb,p)
v_o = sicro.GetMoistAirVolume(T_odb,w_o,p)
h_o = sicro.GetMoistAirEnthalpy(T_odb,w_o)/1000
m_o = sys_df["outdoor_air_flow"][8759]/v_o

T_idb = sys_df["T_idb"][8759]
T_iwb = sys_df["T_iwb"][8759]
w_i = sicro.GetHumRatioFromTWetBulb(T_idb,T_iwb,p)
v_i = sicro.GetMoistAirVolume(T_idb,w_i,p)
h_i = sicro.GetMoistAirEnthalpy(T_idb,w_i)/1000
m_i = system.parameter("supply_air_flow").value/v_i

T_supply = sys_df["T_supply"][8759]
w_supply = sys_df["w_supply"][8759]/1000
v_supply = sicro.GetMoistAirVolume(T_supply,w_supply,p)
h_supply = sicro.GetMoistAirEnthalpy(T_supply,w_supply)/1000
m_supply = system.parameter("supply_air_flow").value/v_supply

# Coil output
T_co = T_supply - sys_df["fan_power"][8759]/(system.parameter("supply_air_flow").value * system.props["RHO_A"]*system.props["C_PA"])
w_co = w_supply
v_co = sicro.GetMoistAirVolume(T_co,w_co,p)
h_co = sicro.GetMoistAirEnthalpy(T_co,w_co)/1000
m_co = system.parameter("supply_air_flow").value/v_co
print("Coil output HR(%): ",sicro.GetRelHumFromHumRatio(T_co,w_co,p)*100)

# Return air
T_ra = space_df["temperature"][8759]
w_ra = space_df["abs_humidity"][8759]/1000
v_ra = sicro.GetMoistAirVolume(T_ra,w_ra,p)
h_ra = sicro.GetMoistAirEnthalpy(T_ra,w_ra)/1000
m_ra = (system.parameter("supply_air_flow").value-0.09439)/v_ra


excel = pd.DataFrame()
excel["OA"] = [T_odb,w_o,v_o*1000,h_o,m_o]
excel["MA"] = [T_idb,w_i,v_i*1000,h_i,m_i]
excel["CO"] = [T_co,w_co,v_co*1000,h_co,m_co]
excel["SA"] = [T_supply,w_supply,v_supply*1000,h_supply,m_supply]
excel["RA"] = [T_ra,w_ra,v_ra*1000,h_ra,m_ra]

excel.to_clipboard()
excel
```
