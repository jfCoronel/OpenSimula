## Calculator component Test


```python
import opensimula as osm

case_dict = {
    "name": "Calculator Test",
    "time_step": 3600,
    "n_time_steps": 8760,
    "initial_time": "01/01/2001 00:00:00",
    "components": [
        {
            "type": "File_met",
            "name": "Denver",
            "file_type": "TMY3",
            "file_name": "../../mets/WD100.tmy3"
        },{
            "type": "Calculator",
            "name": "Cambiar unidad",
            "input_variables": ["T = Denver.temperature", "w = Denver.abs_humidity"],
            "output_variables": ["T_F","W_kg", "degree_hour_20","cos"],
            "output_units": ["ºF","kg/kg a.s.","ºC"," "],
            "output_expressions": ["T *9/5 + 32", "w / 1000", "0 if T>20 else 20-T","math.cos(math.pi)"],
        }
    ]
}


sim = osm.Simulation()
pro = sim.new_project("pro")
pro.read_dict(case_dict)
```

```python
pro
```

```python
pro.component("Cambiar unidad")
```

```python
pro.simulate()
```

```python
pro.component("Cambiar unidad").variable_dataframe()
```

```python
sim.plot(pro.dates(),[pro.component("Cambiar unidad").variable("T"),pro.component("Cambiar unidad").variable("degree_hour_20")])
```
