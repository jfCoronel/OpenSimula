## Personalizar la visualización Plotly

```python
# Obtener la figura y personalizarla
fig = env_3D.show_plotly(polygons_type="initial")

# Cambiar título
fig.update_layout(title="Mi Edificio - Vista 3D")

# Cambiar posición de cámara
fig.update_layout(
    scene_camera=dict(
        eye=dict(x=2, y=-2, z=1.5)
    )
)

# Cambiar altura
fig.update_layout(height=900)

fig.show()
```
